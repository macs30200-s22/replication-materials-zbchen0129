```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import string

import matplotlib.pyplot as plt
% matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("antiva_dataset.csv", index_col=0)
df.head()
```

```python
text = " ".join(review for review in df.tweet)

# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(list(string.punctuation) + ['people', 'think', 'know', "https", "anti", 'antivaxxer', 'antivaxxers', 'anti-vaxxers', 'anti-vaxxer', "vaccine", "vaccinated", "vaccination", "vax", 'covid','covid-19', 'covid19', "vaccines", 'vaxxer', 'vaxxers', 't', 'co', 'pandemic', 'anti-vaccine', 'amp'])



wordcloud = WordCloud(stopwords=stopwords, background_color="white", min_word_length=4, collocation_threshold=3).generate_from_text(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
